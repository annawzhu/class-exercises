````python
# Lasso
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer


grid = {
    'estimator__alpha': np.logspace(-3, -1, 100)
}

# A note on pipelines
# pipeline as essentially a 'wrapper' to make a group of transformers with a final estimator look and behave 
# (programmatically) like a single estimator. 
# So when you call 'fit' on the pipeline, it takes whatever data you pass it, 
# runs it through each of the transformers (SimpleImputer etc) in turn, 
# and then passes it to the 'fit' method of the estimator in the pipeline - just the same
# as if you called 'fit' directly on an OLS estimator instance.

# A benefit of using pipeline is.. 
# as it provides a wrapper that makes it easier to wrap up all your data processing in one 
# 'interface' - it's less tempting to just run the SimpleImputer once on the whole dataset 
# and then do train/test splits etc. 
# The cross_validate function and GridSearchCV class then contain the 
# logic to do the required splits for cross-validation etc - and pass each split into the pipeline.
# As for the imputation on the test data - that happens implicitly as when predict is 
# called on the pipeline for the test data, it will be run through the imputation step then.

pipe = Pipeline([
    ('imputer', SimpleImputer(missing_values=np.nan, strategy='median')),
    ('standardise', StandardScaler()),
    ('estimator', Lasso())
])

# Setup the GridSearch
# This time compare scores for both R-squared and MSE
# NOTE: We actually use negative mean squared error, as scikit-learn always tries to maximise the score 
# NOTE: We're actually going to use two types of scores: R-sq and negative mean squared error 

# later on in the code, when we do LOO, as will use -MSE as our 'primary' scoring method, 
# so the results obtained are comparable with results below

# note, here, we begin with 3-fold inner and 3-fold outer CV
# STUDENTS: If you wanted to change to 5-fold outer and 2-fold inner - what parts of the code would you change?



# Note 'scoring' below shows two metrics (enabling us to present the graphs of both R2 and MSE)
# but GridSearchCV still needs to have a scoring metric on which to calculate the 'best' parameters found

# A note on what GridSearchCV is doing 'internally': it is running through each 'grid point' (combination of parameters - 
# in this case just lambda/alpha), then fitting (actually running the whole pipeline) the training data, then predicting the 
# test data based on this fit, and calculating whatever scoring metric is specified based on this. 
# In this case because we specified both NMSE and R2 it will calculate both of these scores at each grid point. 
# So we can go back and plot each alpha vs either MSE or R2. 
# At the end of the GridSearchCV the 'best' estimator parameters will be chosen based only on (N)MSE though.


scoring = ['r2', 'neg_mean_squared_error']

grid_search_cv = GridSearchCV(estimator=pipe, param_grid=grid, cv=3, return_train_score=True, scoring=scoring, 
                              refit='neg_mean_squared_error')



# STUDENTS: to change to 2-fold inner....
# grid_search_cv = GridSearchCV(estimator=pipe, param_grid=grid, cv=2, return_train_score=True, scoring=scoring, 
#                               refit='neg_mean_squared_error')
````

````python
# Nested CV
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_validate 

cv_result_cv = cross_validate(estimator=grid_search_cv, X=X, y=y, cv=3, return_train_score=True, return_estimator=True,
                           scoring=scoring)

# What results "show up in cv_result_cv"? 
# grid_search_cv is the initialised GridSearchCV object which we then pass into  e.g. `cross_validate`. 
# It has a range of methods and attributes, see the documentation
# (https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html)

# On a 'fit' GridSearchCV object (ie not this one at this point in the code) you can access attributes such as
# - `best_estimator_`
# - `best_score_`
# - `best_params_`
# but exactly what's accessible depends on the state of the object

# STUDENTS: to change to 5-fold outer....
# cv_result_cv = cross_validate(estimator=grid_search_cv, X=X, y=y, cv=5, return_train_score=True, return_estimator=True,
#                            scoring=scoring)
````

````python
# R^2
# Note: these are the average scores across the 3 outer folds..using the 'best' lambda

# the keys returned in cv_results_cv are:
# - fit_time
# - score_time
# - estimator
# - test_r2
# - train_r2
# - test_neg_mean_squared_error
# - train_neg_mean_squared_error
# Most of these are arrays with one value from each CV fold (3 in this case)
# Note that these are configurable/change depending on usage (e.g. estimator is included because we asked for it, 
# same for train score. 
# Obviously the score names changes depending on which scores we've asked for.)
# These aren't 'by default' the results from the 'best' CVs - that behaviour is provided by the GridSearchCV we pass in - 
# `cross_validate` purely runs the CV process and collates the results.

test_score = cv_result_cv['test_r2']
train_score = cv_result_cv['train_r2']
plt.bar(x=range(2), 
        height = [np.mean(train_score), np.mean(test_score)], 
        tick_label=['Train', 'Test'], 
        yerr=[np.std(train_score), np.std(test_score)])
plt.ylabel(r'$R^2$')
plt.title('"Outer" cross-validation scores ($R^2$)')
plt.show()
````

````python
# MSE
# Note: these are the average scores across the 3 outer folds..using the 'best' lambda

test_score = cv_result_cv['test_neg_mean_squared_error']
train_score = cv_result_cv['train_neg_mean_squared_error']
plt.bar(x=range(2), 
        height = np.multiply(-1, [np.mean(train_score), np.mean(test_score)]), 
        tick_label=['Train', 'Test'], 
        yerr=[np.std(train_score), np.std(test_score)])
plt.ylabel(r'MSE')
plt.title('"Outer" cross-validation scores (MSE)')
plt.show()
````

````python
# STUDENTS: in above figures, why do you think the outer train score always show 
# superior performance of the model compared to the outer test score?
# i.e. Train has: higher R-sq and lower MSE...
# Is that what you would expect? Why?
# If Test score is substantially "worse-off" than train score - what does that suggest?
# A: overfitting

# Note: we only show 'one' average score from the three outer scores from the 3 outer folds
# we could show three outer scores but since the purpose of the outer folds
# is to test the 'overall' generalisability of the model, showing one average is fine

# by contrast, below we show you 'three' separate sets of scores from the 3 folds from the outer folds
# because we are interested in showing how each lambda performed in each of the folds
# each point on the line is the average of the 3 inner folds
# each line is for an outer fold
````

````python
# Let's look at the 'internal' test scores for each alpha (lambda) (R^2)

fig, ax = plt.subplots()
ax.set_xscale(value='log')

for estimator in cv_result_cv['estimator']:
    result = estimator.cv_results_
    err = plt.errorbar(x=result['param_estimator__alpha'].data, y=result['mean_test_r2'], yerr=result['std_test_r2'], label='Test')
    plt.errorbar(x=result['param_estimator__alpha'].data, y=result['mean_train_r2'], yerr=result['std_train_r2'], label='Train', 
                 linestyle='dashed', color=err[0].get_color())
    plt.axvline(x=[estimator.best_params_['estimator__alpha'],], color=err[0].get_color())
plt.xlabel('alpha/lambda')
plt.ylabel(r'$R^2$')
plt.title('"Inner" test vs train scores ($R^2$)')
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# Students: why are there 6 lines in the above figure?
# first: are these scores being generated on the 'inner' or 'outer' folds of data
# A: three outer folds - so each colour represents the GridSearchCV 
# run in one of the 3 'outer' folds

# but hold on - didn't we also do 3 inner folds (for each separate outer fold), what happened to them? 
# A: indeed, we did 3-fold "inner" CV and for each inner fold split of train data, we did hyperparameter tuning. 
# But above, for each outer split, we then look at the 'average/stdev' of the (3) 'inner' scores 
# at each tested value of alpha/lambda, which are the actual points plotted

# Put another way, each 'train' line is the average of the 'inner' 3-fold 'train' data outcomes at 
# each lambda, and the same for test (average of the innner 3-fold test runs on each 'outer' run) 

# Students: to the right of the vertical lines "the best lambdas" 
# why do you think the R-sq rapidly declines?
# A: underfitting

# Students: why do you think you get divergence in train and test scores for 
# 'small' values of lambda?
# A: overfitting. small values of lambda means little regularisation/penalisation
# thus you're including many features in the model.

# Students: why do you think you get overlap in train and test scores for 
# 'large' values of lambda? 
# A: large values of lambdas mean that you're penalising heavily e.g. 
# in lasso - you're dropping many features. This is leading to "under-fitting"
# which is bad in both train and test cases. Notice how R-sq is very low.

# Students: why do you think there's difference between the orange test, 
# green test and the blue test scores? 
# A: this is just due to the different data sets created by the splitting, 
# and that this is totally expected 
# (in fact you would be surprised if different splits produced identical results)
# However if there were large difference seen then this might suggest 
# something going on with the data
# (e.g. perhaps data ordered in some way and it could be the data were not shuffled) 
````

````python
# Let's look at the 'internal' test scores for each alpha (lambda) (MSE)

fig, ax = plt.subplots()
ax.set_xscale(value='log')

for estimator in cv_result_cv['estimator']:
    result = estimator.cv_results_
    err = plt.errorbar(x=result['param_estimator__alpha'].data, y=np.multiply(-1, result['mean_test_neg_mean_squared_error']), 
                       yerr=result['std_test_neg_mean_squared_error'], label='Test')
    plt.errorbar(x=result['param_estimator__alpha'].data, y=np.multiply(-1, result['mean_train_neg_mean_squared_error']), 
                 yerr=result['std_train_neg_mean_squared_error'], label='Train', linestyle='dashed', color=err[0].get_color())
    plt.axvline(x=[estimator.best_params_['estimator__alpha'],], color=err[0].get_color())
plt.xlabel('alpha/lambda')
plt.ylabel(r'MSE')
plt.title('"Inner" test vs train scores (MSE)')
plt.legend()
plt.tight_layout()
plt.show()
````
