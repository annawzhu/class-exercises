```python
# Plot ViolentCrimesPerPop vs PctFam2Par (line graph)
# first we need to bin the data, and calculate the required stats
import numpy as np
from scipy import stats

num_bins = 50

mean, bin_edges, _ = stats.binned_statistic(x=df['PctFam2Par'], values=df['ViolentCrimesPerPop'], bins=num_bins, statistic='mean')
stdev, bin_edges, _ = stats.binned_statistic(x=df['PctFam2Par'], values=df['ViolentCrimesPerPop'], bins=num_bins, statistic='std')
count, bin_edges, _ = stats.binned_statistic(x=df['PctFam2Par'], values=df['ViolentCrimesPerPop'], bins=num_bins, statistic='count')

sem = stdev / np.sqrt(count)
bin_centres = (bin_edges[0:-1] + bin_edges[1:]) / 2.0

# now plot with error bars
plt.errorbar(bin_centres, mean, yerr=2*sem)
plt.ylabel('ViolentCrimesPerPop')
plt.xlabel('PctFam2Par')
plt.plot()

# Now recreate this plot, using the variable you chose above
```

```python
# This cell uses ipywidgets to display an interactive form of the graph above.
# Use the slider generated to see what happens to the graph as the bin size is changed

import ipywidgets as widgets

def line_graph_with_bins(num_bins=50):
    mean, bin_edges, _ = stats.binned_statistic(x=df['PctFam2Par'], values=df['ViolentCrimesPerPop'], bins=num_bins, statistic='mean')
    stdev, bin_edges, _ = stats.binned_statistic(x=df['PctFam2Par'], values=df['ViolentCrimesPerPop'], bins=num_bins, statistic='std')
    count, bin_edges, _ = stats.binned_statistic(x=df['PctFam2Par'], values=df['ViolentCrimesPerPop'], bins=num_bins, statistic='count')

    sem = stdev / np.sqrt(count)
    bin_centres = (bin_edges[0:-1] + bin_edges[1:]) / 2.0

    # now plot with error bars
    plt.errorbar(bin_centres, mean, yerr=2*sem)
    plt.ylabel('ViolentCrimesPerPop')
    plt.xlabel('PctFam2Par')
    plt.plot()

widgets.interact(line_graph_with_bins, num_bins=widgets.IntSlider(min=1, max=100, value=50));
```

```python
# Plot histogram of ViolentCrimesPerPop
def histogram(num_bins=5):
    plt.hist(x=df['ViolentCrimesPerPop'], bins=num_bins)
    plt.xlabel('ViolentCrimesPerPop')
    plt.ylabel('Count')
    plt.plot()

# Use the slider to explore the effect of bin size
widgets.interact(histogram, num_bins=widgets.IntSlider(min=1, max=20, value=5))
```

```python
# Handling missing variables
# Some variables have missing data
# e.g.
df.describe()['LemasPctPolicOnPatr']
```

```python
# How do we handle this? Several possibilities.
# Drop missing data
# Unfortunately this drops any rows with any missing data - so we end up with significantly less data to work with
df.dropna().describe()
```

```python
# Impute values for missing data - to a constant
df.fillna(0).describe()['LemasPctPolicOnPatr']
```

```python
# Impute values for missing data - to mean
df.fillna(df.mean()).describe()['LemasPctPolicOnPatr']
```

```python
# Impute values for missing data - to median
df.fillna(df.median()).describe()['LemasPctPolicOnPatr']
```
